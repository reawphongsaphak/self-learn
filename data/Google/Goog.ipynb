```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r"/Users/phongsaphak/Documents/data_set/GOOG.csv")
df
```

```python
df["Price"] = (df["Open"]+df["High"])/2
df["Date"] = pd.to_datetime(df["Date"])
df.index = df["Date"]
df.drop(["Open","Low","High","Close","Adj Close","Volume","Date"],axis=1,inplace=True)
df.head()
```

```python
df.plot(figsize=(20,8))
plt.xlabel("year")
plt.ylabel("price (dollars)")
plt.title("Google stocks price")
plt.grid()
plt.show()
```
