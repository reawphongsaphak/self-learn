```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample multi-index DataFrame
arrays = [['A', 'A', 'B', 'B'], [1, 2, 1, 2]]
index = pd.MultiIndex.from_arrays(arrays, names=('First', 'Second'))
df = pd.DataFrame({'Values': [10, 20, 30, 40]}, index=index)

df
```

```python
# Plotting
# If you want to plot each level of the index separately:
for name, group in df.groupby(level='First'):
    group.plot(kind='bar', legend=True, title=name)

plt.show()

# If you want to plot both levels of the index together:
df.unstack().plot(kind='bar', legend=True)
plt.show()
```
